Prove if high humidity (60% onwards = TRUE) & cloudy days (2 to 8) attributes to rain from above. Using Conditional Probability, that the two attributes overlap will contribute to % probability of rain. 

```python
len(df)
```

```python
#create boolean column for 60% humidity
df['60%humid'] = np.where(df['Humidity9am'] >= 60, 1,0)

#create boolean column for Cloud count above 6
df['highcloud'] = np.where(df['Cloud9am'] >= 7, 1,0)

#index column
df['count'] = 1

#create probability dataframe from above
df = df[['60%humid', 'highcloud','count']]
df.head()

```

```python
#create a pivot table from above
pd.pivot_table(df, values='count',
              index=['60%humid'],
              columns=['highcloud'],
              aggfunc=np.size,fill_value=0)
```

Now we have the data we need for Probability Formula;
where 
    P(A) = 60%humid;
    P(B) = highcloud;
    P(A|B) = probability of rain given A and B;
